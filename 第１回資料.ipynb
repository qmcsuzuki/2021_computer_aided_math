## if 文
if 文は「条件が成り立つ場合に以下を実行する」という内容を実現できる。
if 文の文法は以下の形である。
ここで、「条件式が真のときに実行する内容」の行が右にずれている（インデントされている）が、これも文法の一部である。

インデントを入れるには Tab キーを押せばよい。また、１回インデントを入れた後で改行すると、colab のエディターが気を利かせてインデントを自動で挿入してくれる。なので必要に応じて自分でインデントを削除すること。

```
if (条件式):
    条件式が真のときに実行する内容
```

ここで条件式の例として以下をあげる
- $a$が正 `a > 0`
- $a$が$1$と等しい `a == 1` （`=` だと代入になることに注意！！）
- $a$が$3$の倍数($3$で割った余りがゼロ) `a%3 == 0`
- $a$が正かつ$3$の倍数 `a > 0 and a%3==0`
- $a$が正または$3$の倍数 `a > 0 or a%3==0`




````python
a,b = 1,2
if a > 0:   # もし a が 正なら
    a += 1  # a に 1 を足し、
    b += 1  # さらに b に 1 を足す
print(a,b)
````

「ここまでの条件が成り立たず、かつ次の条件が成り立つ場合」は `elif` (C言語における `else if`)で、
「どの条件にも当てはまらない場合」は `else` であらわす。

`elif`, `else` についても同様に、インデントされた部分のみを実行する。

````python
a,b = 1,2
if a % 3 == 0:   # もし a が 3の倍数なら
    a += 1  # a に 1 を足し、
    b += 1  # さらに b に 1 を足す
else:       # もし if 文の条件式が成り立たない場合（つまり a が 3 の倍数でない場合、今回はこっち）
    a -= 10 # a から 10 を引き、
    b -= 10 # さらに b から 10 を引く
print(a,b)  # この行はインデントされていないので、if 文の条件式の真偽によらずに実行される
````

関数と if 文を組み合わせると、「場合分けする関数」を書くことができる。
関数の中に `return` は何個書いてもよい。

関数の中に if 文や for 文を書く場合、ルール通り（もう一段）インデントする。

以下のコードは、負、ゼロ、正の値についてそれぞれ -1,0,1 を返す「符号関数」sgn を定義して、いくつかの関数値を表示している。
（自分で関数を定義する方法はあとで説明する。）


````python
def sgn(x): # 符号関数 sgn を以下で定義
    if x > 0:    # 正なら
        return 1
    elif x == 0: # （正でなく）0 なら （「等しい」という条件をあらわすときは == を使う）
        return 0
    else:        # 上の２条件にあてはまらない場合 
        return -1

print(sgn(-6), sgn(0), sgn(1000))
````

## for 文
for 文は「繰り返し」や、「リストの元すべてについて～」という操作に対応する。
以下のように書くと、リスト `a` の元を頭から走査する。

以下のコードで、 `a += 1` の行が右にずれている（インデントされている）が、これも for 文の文法の一部である。
for 文は、インデントされている部分を繰り返す。

````python
a = [6,7,8,9]
print(a)
for i in a:
    print(i*i)
````

次のコードでは、for 文の下２行がインデントされているので、その２行が繰り返される。

````python
a = [2,4,6]
for i in a:
    i += 1
    print(i)
print(100) # この行はインデントされていないので、for 文の繰り返しが終わった後に（１回だけ）実行される。
````

### range

`range(x)` は、$0$ 以上 $x$ **未満**の列をあらわす。
つまり長さが $x$ である。
たとえば `range(5)` で、列 0,1,2,3,4 を表す。

`range`を使うと、「$x$ 回繰り返す」という操作が簡単に書ける。
たとえば「`a` に 1 を足す」という操作を 5 回繰り返したいときは次のように書く。

````python
a = 0
for i in range(5):
    a += 1
````

また、`range(1,10,2)` とかくと、$1$ からはじまる公差 $2$ の等差数列で、値が $10$ **未満**の列を意味する。

````python
for i in range(1,10,2):
    print(i)
````

`range` は実はリストではない。リストにしたい場合は`list(range(5))`のようにする。

````python
print(range(2,5))

a = list(range(2,5))
print(a)
````

### リストの要素そのものを操作したいとき

「リスト `a` の元すべてを２乗する」としたい場合は、`for i in a:` の形式を使うことはできない。
なぜなら、変数 `i` に入っているのはただの数値だからである（「`a`の何番目か？」のような情報は持っていない）

その代わり、`for i in range(len(a)):` と書くことでリスト `a` の添え字を列挙できる。そこで `a[i] = a[i]*a[i]`とすることでリストの i 番目の要素に `a[i]`の２乗が代入される。

````python
a = list(range(6,10))
print(a) # 6 以上 10 未満の数が順に要素となるリスト
for i in a:
    i = i * i
print(a) # リスト a の各要素は何も変わっていない

````

````python
a = list(range(6,10))
for i in range(len(a)): # len(a) で、リスト a の長さがわかる
    a[i] = a[i] * a[i] # a[i] に、a[i]の二乗を代入する
print(a) # リスト a の各要素が二乗された
````

# continue と break

- `break`文 は、for 文を中断する（終わりにする）。
- `continue`文 は、for 文を中断し、次のループの繰り返しの処理を開始させる。

````python
for i in range(5):
    if i == 2:
        break
    print(i)
````

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````

## while 文
while 文は、「条件を満たしているあいだ指示を繰り返す」というタイプの繰り返しを実行できる。

ここで、while の下３行が右にずれている（インデントされている）が、これも for 文の文法の一部である。
while 文は、条件式が満たされている間、インデントされている部分を繰り返す。

以下の関数は、「２で割れるあいだは割り続ける」という指示により、
正の数 x が 2 で何回割れるかを計算する関数である。

たとえば `f(12)` の計算中では、$x$ の値が $12,6,3$ と変化し、２で割れなくなったので
`while`ループから脱出している。

````python
def f(x):
    count = 0
    while x%2 == 0: # 記号 % は「余り」を表す
        x //= 2
        count += 1
    return count

for x in [1,2,3,4,8,12]:
    print(f(x))
````

while 文では、実装を失敗して「無限ループ」になってしまうことがよくある。
以下のコードは無限ループである。

実行が数秒で終わらない場合、無限ループになってしまった可能性がある。
まずセルの左の「停止ボタン」を押してセルの実行を中止しよう。
そのあと落ち着いてコードを見直そう。

````python
def f(x):
    count = 0
    while x%2 == 0:
        # うっかり x を 2 で割るのを忘れてしまった。
        count += 1
    return count

f(4)
````

リストを返す関数 $f(n)$ を作り、$f(10^{5})$ を出力せよ。

例えば `f(8)` の返り値は `[1,2,3,5,8]`となる。

以下の欄で試してみよう。


````python
# ここに書いてみよう






````

**解答例**

````python
def f(n):
    a,b = 1,2
    res = []
    while a <= n:
        res.append(a)
        a,b = b,a+b # python では「同時に代入する」ができる
    return res

print(f(10**5))
````

**解答例２**

Python では、`a[-1]` と書くとリスト `a` の一番右の要素を取り出すことができる。
`a[-2]` はその左の要素になる。


````python
def Fib(n):
    if n == 1:
        return [1]
    a = [1,2]
    while a[-1] + a[-2] <= n:
        a.append(a[-1]+a[-2])
    return a

print(Fib(10**5))
````

## 練習問題４(Fizzbuzz)

関数 $f(x)$ を以下で定める

- 数字 $x$ を返す
- ただし 3で割り切れるときは数字の代わりに「Fizz」を返す
- ただし 5で割り切れるときは数字の代わりに「Buzz」を返す
- 両方で割り切れるときは数字の代わりに「FizzBuzz」を返す

正整数 $n$ を受け取って、
$f(1), \dots, f(n)$ の答えを保存した長さ$n$ のリストを返す関数 $g$ を作ってください。

たとえば $n=18$ のときは $g(18)$ は以下のようになります。

`[1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, 14, FizzBuzz, 16, 17, Fizz]`




````python
#ここに書いてみよう









````

**解答例**

````python
def f(i):
    if i % 15 == 0:
        return "FizzBuzz"
    elif i % 3 == 0:
        return "Fizz"
    elif i % 5 == 0:
        return "Buzz"
    else:
        return i

def g(n):
    a = []
    for i in range(1,n+1):
        a.append(f(i))
    return a

print(g(18))
````

**解答例２**

`list(range(1,n+1))` で初期化する方針。

````python
def FizzBuzz(n):
    a = list(range(1,n+1))
    for i in range(1,n+1): # これは数。a の添え字は i-1 になる
        if i%15 == 0:
            a[i-1] = "FizzBuzz"
        elif i%3 == 0:
            a[i-1] = "Fizz"
        elif i%5 == 0:
            a[i-1] = "Buzz"
    return a

print(FizzBuzz(18))
````
