```python
def solve(n,v):
    def print_board():
        for i in b:
            print("".join(map(str,i)))
        print()
    
    def check(i,j):
        for ii,jj in res:
            if abs(i-ii) <= 2 and abs(j-jj) <= 2:
                return False
        return True
 
    def f(k,m):
        nonlocal cnt
        if m == 0:
            cnt += 1
            #print_board()
            return
        if k + m > n*n:
            return
        
        i,j = divmod(k,n)
        if check(i,j):
            res.append((i,j))
            b[i][j] = 1
            f(k+1,m-1)
            res.pop()
            b[i][j] = 0
        f(k+1,m)

    cnt = 0
    b = [[0]*n for _ in range(n)]
    res = []
    f(0,v)
    return cnt


cnt = solve(8,8)
print(cnt)
```
